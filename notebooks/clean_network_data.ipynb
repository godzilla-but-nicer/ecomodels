 Remove trailing commas from adjacency matrices

 I added trailing commas when saving the networks in Rust. This fucks up
 numpy's ability to load the files and therefore my ability to work with the
 networks as an ensamble. Here I will remove these trailing commas.

# Load adjacency matrix as strings

```python
with open('../data/complete/1_adjmat_10_tester.csv', 'r') as fam:
    lines = fam.readlines()

print(lines)
```

# Remove trailing comma from string

```python
clean_lines = []
for line in lines:
    sl = line.strip()
    if sl[-1] == ',':
        cl = sl[:-1] + '\n'
    else:
        cl = sl + '\n'
    clean_lines.append(cl)

print(clean_lines)
```

# Write new lines

```python
with open('../data/complete/1_adjmat_10_test_out.csv', 'w') as oam:
    for line in clean_lines:
        oam.write(line)

with open('../data/complete/1_adjmat_10_test_out.csv', 'r') as fam:
    lines = fam.readlines()

print(lines)
```

# Let's do it

```python
from glob import glob
am_files = glob('../data/complete/*/*/*_adjmat_10.csv')
for am_file in am_files:
    # open old network
    with open(am_file, 'r') as fam:
        lines = fam.readlines()
    
    # remove trailing commas
    clean_lines = []
    for line in lines:
        sl = line.strip()
        if sl[-1] == ',':
            cl = sl[:-1] + '\n'
        else:
            cl = sl + '\n'
        clean_lines.append(cl)

    # write cleaned files
    with open(am_file, 'w') as oam:
        for line in clean_lines:
            oam.write(line)
```
